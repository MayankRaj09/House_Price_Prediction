```python
dataset['mainroad'] = dataset['mainroad'].astype('category')
dataset['mainroad'] = dataset['mainroad'].cat.codes

dataset['guestroom'] = dataset['guestroom'].astype('category')
dataset['guestroom'] = dataset['guestroom'].cat.codes

dataset['basement'] = dataset['basement'].astype('category')
dataset['basement'] = dataset['basement'].cat.codes

dataset['hotwaterheating'] = dataset['hotwaterheating'].astype('category')
dataset['hotwaterheating'] = dataset['hotwaterheating'].cat.codes
dataset['airconditioning'] = dataset['airconditioning'].astype('category')
dataset['airconditioning'] = dataset['airconditioning'].cat.codes

dataset['prefarea'] = dataset['prefarea'].astype('category')
dataset['prefarea'] = dataset['prefarea'].cat.codes

code_mapping_furniture = {'unfurnished':0, 'semi-furnished':1, 'furnished':2}
dataset['furnishingstatus'] = dataset['furnishingstatus'].astype('category')
dataset['furnishingstatus'] = dataset['furnishingstatus'].map(code_mapping_furniture)

dataset
```

```python
Q1=dataset.quantile(0.25)
Q3=dataset.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
#False value here represent data are valid and true represent the outlier value
print(dataset < (Q1 - 1.5 * IQR)) or (dataset > (Q3 + 1.5 * IQR))
```

```python
dataset_new=dataset[~((dataset<(Q1 - 1.5*IQR))| (dataset>(Q3+1.5*IQR))).any(axis=1)]
dataset_new.shape
```

```python
#correlations between quality and different columns
correlations=dataset_new.corr()['price'].drop('price').drop('mainroad').drop('guestroom').drop('hotwaterheating').drop('prefarea')
print(correlations)
```

```python
x=dataset.drop(columns='price')
y=dataset['price']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
print("Done")
```

```python
x_train
```

```python
x_test
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
c=model.intercept_
m=model.coef_
print("Model intercept:-",c)
print("Model Coeffiecient:-",m)

```

```python
train_pred=model.predict(x_train)
print("Done")

```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train, train_pred)
plt.xlabel("Actual result")
plt.ylabel("Predicted result")
x_point = np.array([0,14000000])
y_point = np.array([0,14000000])
# max value of y is around 13 million
plt.plot(x_point, y_point, c = 'r')
print("RESULT WITH TRAINED DATA")
print("Number of data train: ", len(x_train))
plt.show()
```

```python
test_pred=model.predict(x_test)
print("Done")
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, test_pred)
plt.xlabel("Actual result")
plt.ylabel("Predicted result")
x_point = np.array([0,14000000])
y_point = np.array([0,14000000])
# max value of y is around 13 million
plt.plot(x_point, y_point, c = 'r')
print("RESULT WITH TRAINED DATA")
print("Number of data train: ", len(x_train))
plt.show()
```

```python
sns.distplot((y_test-test_pred),bins=50)
```

```python
#Result obtained from training the data
from sklearn.metrics import r2_score
r2_score_without_test = r2_score(y_train, train_pred)
print(r2_score_without_test)
```

```python
#Result obtained from testing the data
from sklearn.metrics import r2_score
r2_score_with_test = r2_score(y_test,test_pred)
print(r2_score_with_test)
```
